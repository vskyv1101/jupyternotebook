<h1>작업형 문제1 답안</h1>

```python
import csv
import matplotlib.pyplot as plt

f = open('age_0922.csv')
data = csv.reader(f)

header = next(data)

result = []
for row in data:
    if '신도림' in row[0]:
        for i in row[3:]:
            result.append(int(i))
            
f.close()

plt.plot(result)
plt.xlabel('Age')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

<h1>작업형 문제2 답안</h1>

```python
import csv
import matplotlib.pyplot as plt

f = open('age_0922.csv')
data = csv.reader(f)

header = next(data)

name = input('인구 구조가 알고 싶은 지역의(읍면동 단위)을 입력해 주세요: ')

result = []
for row in data:
    if name in row[0]:
        for i in row[3:]:
            result.append(int(i))
            
f.close()

plt.rc('font', family='Malgun Gothic')
plt.plot(result)
plt.title(name + ' 지역의 연령에 따른 인구수')
plt.xlabel('Age')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

<h1>작업형 문제 답안3</h1>

```python
import csv
import matplotlib.pyplot as plt

f = open('age_0922.csv')
data = csv.reader(f)

header = next(data)

name = input('찾고 싶은 지역의 이름을 알려주세요: ')

result = []
for row in data:
    if name in row[0]:
        for i in row[3:]:
            result.append(int(i))
        
f.close()

plt.figure(figsize=(10, 2))
plt.rc('font', family='Malgun Gothic')
plt.bar(range(len(result)), result)
plt.title(name + ' 지역의 연령에 따른 인구수')
plt.xlabel('Age')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

<h1>작업형 문제4 답안</h1>

```python
import csv
import matplotlib.pyplot as plt

f = open('gender_0922.csv')
data = csv.reader(f)

header = next(data)

name = input('찾고 싶은 지역의 이름을 알려주세요: ')

male = []
female = []
for row in data:
    if name in row[0]:
        for i in row[3:104]:
            male.append(-int(i))
        for i in row[106:207]:
            female.append(int(i))

f.close()

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.barh(range(len(male)), male, color='blue', label='남성')
plt.barh(range(len(female)), female, color='red', label='여성')
plt.title(name + ' 지역의 남녀 연령에 따른 인구수')
plt.xlabel('Population')
plt.ylabel('Age')
plt.grid(True)
plt.legend()
plt.show()
```

<h1>작업형 문제5 답안</h1)

```python
import csv
import matplotlib.pyplot as plt

f = open('gender_0922.csv')
data = csv.reader(f)

header = next(data)

name = input('찾고 싶은 지역의 이름을 알려주세요: ')

for row in data:
    if name in row[0]:
        male = 0
        female = 0
        for i in row[3:104]:
            i = i.replace(',', '')
            male = male + int(i)
        for i in row[106:207]:
            i = i.replace(',', '')
            female = female + int(i)
        break

f.close()

labels = ['남성', '여성']
colors = ['skyblue', 'pink']

plt.rc('font', family='Malgun Gothic')
plt.title(name + ' 지역의 남녀 비율')
plt.pie([male, female], labels=labels, colors=colors, autopct='%.2f%%', startangle=90)
plt.show()
```
