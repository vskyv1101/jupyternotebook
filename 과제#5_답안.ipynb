<h1>작업형 문제1 답안</h1>

<h4>Step 1) 위키백과에서 메달 정보 갖고오기</h4>

```python
import pandas as pd

df = pd.read_html('https://ko.wikipedia.org/wiki/%EC%98%AC%EB%A6%BC%ED%94%BD_%EB%A9%94%EB%8B%AC_%EC%A7%91%EA%B3%84')
df.head(3)
```

<h4>Step 2) '국가 (IOC 코드)'를 인덱스로 지정하기</h4>

```python
df2 = df[0].set_index('국가 (IOC 코드)')
df2.head(3)
```

<h4>Step 3) 하계 메달 기록 데이터만 갖고 오기</h4>

```python
summer = df2.iloc[:, :5]
# iloc은 integer location의 약어로, 데이터 프레임의 행이나 열의 순서를 나타내는 정수로 특정 값을 추출합니다.
summer.head(3)
```

<h4>Step 4) 컬럼 이름을 '경기수', '금', '은', '동', '합계'로 변경하기</h4>

```python
summer.columns = ['경수기', '금', '은', '동', '합계']
summer.head(3)
```

<h4>Step 5) 컬럼 '금'을 기준으로 내림차순으로 정렬하기</h4>

```python
summer = summer.sort_values('금', ascending=False)
summer.head(3)
```

<h4>Step 6) 엑셀 파일로 저장하기</h4>

```python
summer.to_excel('하계올릭픽메달.xlsx')
```

<h1>작업형 문제2 답안</h1>

<h4>Step 1) age_0922.csv 파일에서 데이터 읽어 오기</h4>

```python
import pandas as pd

df = pd.read_csv('age_0922.csv', encoding='cp949')
df.head(3)
```

<h4>Step 2) 인덱스를 '행정구역'으로 설정하기</h4>

```python
df = df.set_index('행정구역')
df.head(3)
```

<h4>Step 3) 자료형 확인하기</h4>

```python
df.dtypes
```

<h4>Step 4) 쉼표(,) 제거하기</h4>

```python
df = df.replace(',', '', regex=True)
df.head(3)
```

<h4>Step 5) 정수형 자료형으로 변환하기</h4>

```python
df = df.apply(pd.to_numeric)
df.head(3)
```

<h4>Step 6) 전체 데이터를 총 인구수로 나누어 비율로 변환하기</h4>

```python
df = df.div(df['2022년08월_계_총인구수'], axis='index')
df.head(3)
```

<h4>Step 7) '2022년08월_계_총인구수', '2022년08월_계_연령구간인구수' 컬럼 제거하기</h4>

```python
del df['2022년08월_계_총인구수'], df['2022년08월_계_연령구간인구수']
df.head(3)
```

<h4>Step 8) 사용자로부터 원하는 지역의 이름 입력 받기</h4>

```python
name = input('원하는 지역의 이름을 입력하세요: ')
```

<h4>Step 9) DataFrame의 인덱스 문자열로부터 원하는 문자열이 포함된 행 찾기</h4>

```python
a = df.index.str.contains(name)
print(a) # 마스크 형태로 추출됩니다.
```

<h4>Step 10) 마스크를 이용하여 데이터 프레임에서 정보 갖고 오기</h4>

```python
df2 = df[a]
df2.head(3)
```

<h4>Step 11) 갖고 온 데이터를 전체 데이터 프레임에서 빼기</h4>

```python
x = df.sub(df2.iloc[0], axis='columns')
x.head(3)
```

<h4>Step 12) 음수를 없애기 위해서 제곱하기</h4>

```python
import numpy as np

y = np.power(x, 2)
y.head(3)
```

<h4>Step 13) 각 지역별로 모든 연령에 대해 계산된 값들을 더하기</h4>

```python
z = np.sum(y, axis='columns')
z.head(10)
```

<h4>Step 14) 0으로 계산된 지역을 제외시키기</h4>

```python
# 일부 행정구역의 연령별 인구수 정보가 전부 0인 경우가 있습니다.
# 이 행정구역들의 경우 z 값이 0으로 계산되기 때문에 제외시켜줍니다.
z = z[z[:] != 0]
z.head(10)
```

<h4>Step 15) 오름차순으로 정렬하고, 가장 비슷한 상위 5개 지역의 인덱스 정보 추출하기</h4>

```python
i = z.sort_values().index[:5]
i
```

<h4>Step 16) 추출한 5개 인덱스 정보로 선 그래프 그리기</h4>

```python
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
df.loc[i].T.plot(figsize=(10, 4))
plt.xlabel('나이')
plt.ylabel('인구 비율')
plt.grid(True)
plt.show()
```
