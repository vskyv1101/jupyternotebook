<h1>작업형 문제1 답안</h1>

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul_0914.csv')
data = csv.reader(f)

header = next(data)

result = []

for row in data:
    if row[4] != '':
        result.append(float(row[4]))
        
f.close()

plt.figure(figsize=(10, 2))
plt.plot(result, 'r')
plt.xlabel('Day')
plt.ylabel('The highest temperature')
plt.show()
```

<h1>작업형 문제4 답안</h1>

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul_0914.csv')
data = csv.reader(f)

header = next(data)

mean = []
low = []
high = []

for row in data:
    if (row[2] != '') and (row[3] != '') and (row[4] != ''):
        if (row[0].split('-')[1] == '12') and (row[0].split('-')[2] == '25'):
            mean.append(float(row[2]))
            low.append(float(row[3]))
            high.append(float(row[4]))
        
f.close()

plt.figure(figsize=(10, 2))
plt.plot(mean, 'k')
plt.plot(low, 'b')
plt.plot(high, 'r')
plt.xlabel('Day (December 25)')
plt.ylabel('The highest temperature')
plt.show()
```

<h1>작업형 문제5 답안</h1>

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul_0914.csv')
data = csv.reader(f)

header = next(data)

mean = []
low = []
high = []

for row in data:
    if (row[2] != '') and (row[3] != '') and (row[4] != ''):
        if (row[0].split('-')[1] == '12') and (row[0].split('-')[2] == '25'):
            mean.append(float(row[2]))
            low.append(float(row[3]))
            high.append(float(row[4]))
        
f.close()

plt.figure(figsize=(10, 2))
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('크리스마스의 기온 변화 그래프')
plt.plot(mean, 'k', label='mean')
plt.plot(low, 'b', label='low')
plt.plot(high, 'r', label='high')
plt.xlabel('Day (December 25)')
plt.ylabel('The highest temperature')
plt.legend()
plt.grid()
plt.show()
```

<h1>작업형 문제6 답안</h1>

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul_0914.csv')
data = csv.reader(f)

header = next(data)

year = []
day = []
mean = []

for row in data:
    y = int(row[0].split('-')[0])
    if (y != 1907) and (y != 1950) and (y != 1951) and (y != 1952) and (y != 1953) and (y != 2022):
        if row[2] != '':
            day.append(float(row[2]))
            if (row[0].split('-')[1] == '12') and (row[0].split('-')[2] == '31'):
                year.append(y)
                mean.append(sum(day)/len(day))
                day = []
        
f.close()

plt.figure(figsize=(10, 4))
plt.rc('font', family='Malgun Gothic')
plt.title('평균 기온 변화 그래프')
plt.plot(year, mean, color='k', marker='d')
plt.xlabel('Year')
plt.ylabel('Average Temperature')
plt.grid()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul_0914.csv')
data = csv.reader(f)

header = next(data)

month = [[], [], [], [], [], [], [], [], [], [], [], []]

for row in data:
    if row[4] != '':
        month[int(row[0].split('-')[1])-1].append(float(row[4]))
        
f.close()

plt.boxplot(month)
plt.show()
```
